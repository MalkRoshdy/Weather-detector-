```python
import numpy as np

X_train = np.load('/content/final_data/X_txt_tfidf_train.npy')
X_test = np.load('/content/final_data/X_txt_tfidf_test.npy')
X_val = np.load('/content/final_data/X_txt_tfidf_val.npy')

y_train = np.load('/content/final_data/y_train.npy')
y_test = np.load('/content/final_data/y_test.npy')
y_val = np.load('/content/final_data/y_val.npy')

print(X_train.shape)
print(y_train.shape)
```

```python
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)

X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

X_val = X_val.reshape(X_val.shape[0], X_val.shape[1], 1)

print(X_train.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D
from tensorflow.keras.layers import MaxPooling1D
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout

model = Sequential()

# CNN
model.add(Conv1D(
    filters=128,
    kernel_size=3,
    activation='relu',
    input_shape=(365,1)
))

model.add(MaxPooling1D(pool_size=2))

model.add(Flatten())

# ANN
model.add(Dense(128, activation='relu'))

model.add(Dropout(0.5))

model.add(Dense(64, activation='relu'))

# Output
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D
from tensorflow.keras.layers import Flatten, Dense, Dropout

model = Sequential()

# CNN
model.add(Conv1D(
    128,
    3,
    activation='relu',
    input_shape=(365,1)
))

model.add(MaxPooling1D(2))

model.add(Flatten())

# ANN
model.add(Dense(128, activation='relu'))

model.add(Dropout(0.5))

model.add(Dense(64, activation='relu'))

# Output for 6 classes
model.add(Dense(6, activation='softmax'))

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True
)

history = model.fit(
    X_train,
    y_train,
    epochs=20,
    batch_size=32,
    validation_data=(X_val, y_val),
    callbacks=[early_stop]
)
```

```python
from sklearn.metrics import f1_score

pred = model.predict(X_test)

pred = pred.argmax(axis=1)

f1 = f1_score(y_test, pred, average='weighted')

print("F1 Score:", f1)
```

```python
X_img_train = np.load('/content/final_data/X_img_train.npy')
X_img_test = np.load('/content/final_data/X_img_test.npy')
X_img_val = np.load('/content/final_data/X_img_val.npy')

print(X_img_train.shape)
```

```python
from tensorflow.keras.layers import Input
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Flatten
from tensorflow.keras.models import Model

image_input = Input(shape=(224,224,3))

x1 = Conv2D(32, (3,3), activation='relu')(image_input)

x1 = MaxPooling2D((2,2))(x1)

x1 = Flatten()(x1)
```

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Concatenate

text_input = Input(shape=(365,))

merged = Concatenate()([x1, text_input])

x = Dense(128, activation='relu')(merged)

x = Dense(64, activation='relu')(x)

output = Dense(6, activation='softmax')(x)

model = Model(inputs=[image_input, text_input], outputs=output)
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
history = model.fit(
    [X_img_train, X_train],
    y_train,

    validation_data=([X_img_val, X_val], y_val),

    epochs=10,
    batch_size=32
)
```

```python
from sklearn.metrics import f1_score

pred = model.predict([X_img_test, X_test])

pred = pred.argmax(axis=1)

f1 = f1_score(y_test, pred, average='weighted')

print("F1 Score:", f1)
```
